```python
inic =''
def estrat(x):
    global inic
    if inic!=x:
        inic = x
        return x
    
    else:
        return np.nan

long = 0
longa =pd.Series()
def rollmeanstrat(serie,longmean=100,shortmean=10):
    global long 
    global longa
    serie =pd.DataFrame(serie)
    if longmean== long:
        lmean =longa
    else:
        lmean= serie.rolling(longmean,min_periods=5).mean()
    
    serie['roll']   =  serie.rolling(shortmean,min_periods=5).mean()-lmean

    
    serie['DECISAO']=np.nan
    
    serie.loc[serie['roll']>0.0,'DECISAO']='COMPRA'
    serie.loc[serie['roll']<=0.0,'DECISAO'] ='VENDA'
    serie['ESTRAT'] =serie['DECISAO'].apply(func=estrat)
    x =returns(serie)
   
    return x
def returns(serie,tipo_oper ='COMPRA'):
    serie['VENDA']= serie.dropna().iloc[:,0].shift(-1)
    serie['LUCRO'] =(serie['VENDA']-serie.iloc[:,0])/(serie.iloc[:,0])
    vals= serie[serie['ESTRAT']==tipo_oper ]
    
    

    def flucro(lucro):
        if lucro>0:
            return 1+lucro
        else:
            return 1-abs(lucro)

    vals['retorno']= vals['LUCRO'].apply(flucro).dropna()
    vals=vals.dropna()
    retorno_total=vals['retorno'].rolling(len(vals)).apply(np.prod, raw=True).dropna()
    return retorno_total
```
